## Question
### a) Commonality  : How many matchs do we have, and how much can common characteristics change the trend ? 

```python
# General Match rate
match = sd['match'].value_counts()

# filtre sur les same race : 
match_race = sd['match'][sd['samerace']==1].value_counts()

# Same age Match with 1 year difference
sd['SameAge'] = (sd['age'] - sd['age_o'] >= -1) & (sd['age'] - sd['age_o'] <= 1)

match_age = sd['match'][sd['SameAge']==True].value_counts()

sd_2 = pd.read_csv('Speed Dating Data.csv', encoding="ISO-8859-1")
sd_double = pd.merge(sd, sd_2, left_on='iid', right_on='pid', suffixes=('_subject', '_partner'))

# Same studies
match_studies = sd_double['match_subject'][sd_double['field_cd_subject']==sd_double['field_cd_partner']].value_counts()

# Same career
match_career = sd_double['match_subject'][sd_double['career_c_subject']==sd_double['career_c_partner']].value_counts()

# Grow-up in the same city
match_from = sd_double['match_subject'][sd_double['zipcode_subject']==sd_double['zipcode_partner']].value_counts()
```

```python
fig, axes = plt.subplots(2,3)
fig.set_size_inches(20,15)

axes[0,0].pie(match, labels=match.index, autopct='%1.1f%%')
axes[0,0].set_title('General match rate', color='white', fontsize=30)

axes[0,1].pie(match_race, labels=match_race.index, autopct='%1.1f%%') 
axes[0,1].set_title('Same race match rate', color='white', fontsize=20)

axes[0,2].pie(match_age, labels=match_age.index, autopct='%1.1f%%') 
axes[0,2].set_title('Same age match rate', color='white', fontsize=20)

axes[1,0].pie(match_studies, labels = match_studies.index, autopct='%1.1f%%')
axes[1,0].set_title('Same field of studies match rate', color='white', fontsize=20)

axes[1,1].pie(match_career, labels = match_career.index, autopct='%1.1f%%')
axes[1,1].set_title('Same intended career match rate', color='white', fontsize=20)

axes[1,2].pie(match_from, labels = match_from.index, autopct='%1.1f%%') 
axes[1,2].set_title('Gowth in the same city match rate', color='white', fontsize=20)

plt.show()
```

### b) Di Caprio gate : is it true that males always choose younger women ? 

```python
#change number to string to improve readability
sd['gender_name'] = sd['gender'].apply(lambda x : 'Female' if x == 0 else 'Male')

sd_match = sd[sd['match'] == 1]
sd_match['dif_age'] = sd_match['age'] - sd_match['age_o']

sd_yes = sd[sd['dec'] == 1]
sd_yes['dif_age'] = sd_yes['age'] - sd_yes['age_o']

```

```python
fig = px.box(sd_yes, x="age", y="dif_age", color='gender_name',
             title="Age difference for yes")
fig.show()
```

```python
fig = px.box(sd_match, x="age", y="dif_age", color='gender_name',
             title="Age difference for match")
fig.show()
```

We can detect that people after 27, man and woman start to prefer younger people. In could be interesting to have more people older to see how the trend is moving. 

## C) self-confidence and actractivity
### Do people with high confidence have more match ? 
### Does more matches = more self-confidence ?

```python
# creation of global rate

sd['self-confidence'] = sd['attr3_1']+sd['sinc3_1']+sd['intel3_1']+sd['fun3_1']+sd['amb3_1']
```

```python
# create an average of match obtained 

sd_match = sd[sd.match ==1]

# match 
nbr_match_per_pax = sd_match.groupby('iid')['match'].sum()
nbr_match_per_pax = nbr_match_per_pax.reset_index()

# date 
nbr_date_per_pax = sd['iid'].value_counts()
nbr_date_per_pax = nbr_date_per_pax.reset_index()

avg_match_tab = nbr_match_per_pax.merge(nbr_date_per_pax, left_on='iid', right_on='index', how='left')
avg_match_tab = avg_match_tab.drop(columns='index')
avg_match_tab['avg_match'] = avg_match_tab['match'] / avg_match_tab['iid_y']

sd_avg = sd.merge(avg_match_tab,left_on='iid', right_on='iid_x', how='left' )
```

```python
fig = px.scatter(sd_avg, x = "self-confidence", y = "avg_match",color = "gender_name",
                 trendline = "ols", title="Self-confidence and match")
fig.show()
```

There isn't real correlation between those 2 variables. We can see it with the really low R2. So there is no correlation between the number of matches and self-confidence in our case. 

```python
# self-confidence after 3 week
sd_avg['self-confidence2'] = sd['attr3_2']+sd['sinc3_2']+sd['intel3_2']+sd['fun3_2']+sd['amb3_2']
sd_avg['self-confidence3'] = sd['attr3_3']+sd['sinc3_3']+sd['intel3_3']+sd['fun3_3']+sd['amb3_3']

sd_avg['evol-self-confidence_1d'] = sd_avg['self-confidence2'] - sd_avg['self-confidence']
sd_avg['evol-self-confidence_3w'] = sd_avg['self-confidence3'] - sd_avg['self-confidence']
```

```python
fig = px.strip(sd_avg, x = "avg_match", y = "evol-self-confidence_1d", color = "gender_name", 
                 title="Self-confidence and match after 1 day", stripmode = "overlay")
fig.show()
```

```python
fig = px.strip(sd_avg, x = "avg_match", y = "evol-self-confidence_3w", color = "gender_name", 
                 title="Self-confidence and match after 3 weeks", stripmode = "overlay")
fig.show()
```

We can't see any significant impact  on the selt-confidence between having a lot of matches and no matches.
